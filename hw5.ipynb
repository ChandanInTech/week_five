```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.MultiDiGraph()

G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
G.add_node('E')

G.add_edge('A', 'B')
G.add_edge('B', 'A')
G.add_edge('B', 'C')
G.add_edge('C', 'A')
G.add_edge('D', 'A')
G.add_edge('D', 'B')
G.add_edge('A', 'E')
G.add_edge('B', 'E')
G.add_edge('E', 'B')
```

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, connectionstyle='arc3, rad = 0.1')
```
